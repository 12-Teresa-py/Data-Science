# Question 39( Sound levels)

```python
sound_level=int(input("Enter a sound level in decibels:"))
if sound_level>130:
  print("Higher than Jackhammer noise")
elif sound_level==130:
  print("Jackhammer noise")
elif sound_level>106:
  print("This sound is between Gas lawnmower and Jackhammer noise")
elif sound_level==106:
  print("Gas lawnmower noise")
elif sound_level>70:
  print("This sound is between Alarm clock and Gas lawnmower noise")
elif sound_level==70:
  print("Alarm clock noise")
elif sound_level>40:
  print("This sound is between Quiet room and Alarm clock noise")
elif sound_level==40:
  print("Quiet room noise")
else:
  print("Lower than Quiet room noise")
```
