# Question 70(Caesar Cipher)

```python
message = input("Enter a message:")
shift = int(input("Enter the shift amount:"))

result = ""
i = 0

while i < len(message):
  char = message[i]

  if char.isupper():
    new_char = chr((ord(char) - ord('A') + shift) % 26 + ord('A'))
    result += new_char

  elif char.islower():
    new_char = chr((ord(char) - ord('a') + shift) % 26 + ord('a'))
    result += new_char

  else:
    result += char

  i += 1

print("Shifted message:", result)
```

# Question 72(Is a String a Palindrome?)

```python
word = input("Enter a word:")

left = 0
right =len(word) - 1
palindrome = True

while left < right:
  if word[left] != word[right]:
    palindrome = False
    break
  left += 1
  right -= 1

if palindrome:
    print(word, "is a palindrome.")
else:
    print(word, "is not a palindrome.")
```

# Question 75(Greatest Common Divisor)

```python
m = int(input("Enter the first positive integer:"))
n = int(input("Enter the second positive integer:"))

if m<n:
  d=m
else:
  d=n

while m % d != 0 or n % d!= 0:
  d = d - 1
  print("The Greatest Common Divisor (GCD) is", d)
```
