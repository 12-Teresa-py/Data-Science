```python
counter = 1
while counter<=3:
  print("*" *(counter*3) )
  counter= counter + 1
```

```python
password_incorrect = True
password = "secret"
while password_incorrect:
  user_pass = input("Enter the password:")
  if user_pass == password:
    print("Welcome back. You are logged in")
    password_incorrect = False
  else:
    print("Password incorrect!!! Please try again")
```

```python
counter = 8
while counter>=2:
  print("*"*(counter*2))
  counter = counter - 1
```

```python
not_entered_zero = True
no_of_items = 0
total = 0
while not_entered_zero:
  val = float(input("Enter a value"))
  if val == 0:
    print(total/no_of_items)
    not_entered_zero = False
  else:
    total = val + total
    no_of_items = no_of_items+1
```
