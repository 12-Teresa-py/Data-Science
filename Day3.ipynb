# Grade converter
It takes the grade and tells the range of marks

```python
grade=input("Enter your grade: ")
if grade == "A":
  print("You scored above 80")
elif grade == "B":
  print("You have scored between 71 and 80 ")
elif grade == "C":
  print("You have scored between 61 and 70")
elif grade == "D":
  print("You have scored between 51 and 60")
elif grade == "E":
  print("You have scored between 41 and 50")
elif grade == "F":
  print("You have scored lower than 41")
else:
  print("The grade you entered is invalid")

```
